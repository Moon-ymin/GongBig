## 부산교통공사 고객의 소리 크롤링

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

import warnings
warnings.filterwarnings('ignore') # 경고문 무시
```

```python
driver = webdriver.Chrome("C://Users/82104/Desktop/chromedriver.exe") # 크롬 드라이버 사용합니다.
driver.implicitly_wait(3)
```

```python
url = 'https://www.humetro.busan.kr/homepage/default/cusvoice/list.do?menu_no=1001020102' # 크롤링할 페이지 주소
driver.get(url)
```

```python
minwonList = []
contentList = []

for i in range(17) :                                                                                                    # 전체 페이지 
    for i in range(10) :                                                                                                # 한 페이지 안에 있는 글 목록이 총 10개
        yesorno = driver.find_element_by_xpath('//*[@id="contents"]/form[2]/div[3]/table/tbody/tr['+str(i+1)+']/td[1]') # 게시글의 공개 여부
        if (yesorno.text) == "O":                                                                                       # 공개된 글만 내용까지 크롤링
            driver.find_element_by_xpath('//*[@id="contents"]/form[2]/div[3]/table/tbody/tr['+str(i+1)+']/td[2]/a').click()
            content = driver.find_element_by_xpath('//*[@id="contents"]/div[4]/div/div[2]/div/div')
            print(content.text) # 크롤링되고 있는지 확인용
            
            article = driver.find_element_by_xpath('//*[@id="contents"]/div[4]/div/div[1]/h3')
            print(article.text)
            minwonList.append(article.text)
            contentList.append(content.text)
            driver.back()
                                                   
        else:
            title = driver.find_element_by_xpath('//*[@id="contents"]/form[2]/div[3]/table/tbody/tr['+str(i+1)+']/td[2]/a')                       
            print(title.text)
            minwonList.append(title.text)
        
    driver.find_element_by_xpath('//*[@id="pagingWrapper"]/ul/li[8]/a/img').click()
driver.quit()
```
