### [그림3] 2020 기준 공시대상 기업 본사 분포

```python
gongsi = pd.read_csv("C://Users/user/Desktop/통합데이터지도공모전/활용데이터/시도별_정보.csv", encoding = 'cp949')
gongsi
```

```python
# 데이터
sido = list(gongsi['시도'])
nums = list(gongsi['공시대상기업'])
index = np.arange(len(sido))

# 시각화
tick_size = 13 # 눈금 폰트 사이즈
axis_label_size = 15 # x축, y축 폰트 사이즈

fig = plt.figure(figsize=(18,8))
ax = fig.add_subplot()

xtick_label_position = list(range(len(sido)))

plt.yticks(fontsize= tick_size)
plt.xticks(xtick_label_position, sido, fontsize=tick_size)

bars = plt.bar(xtick_label_position, nums, alpha = 0.7, linewidth = 3)

for i, b in enumerate(bars):
    ax.text(b.get_x()+b.get_width()*(1/2),b.get_height()+0.3, \
            nums[i],ha='center',fontsize=13)
    
plt.title('공시대상 기업분포',fontsize=18) 
plt.xlabel('시도',fontsize=axis_label_size)
plt.xticks(index, sido, rotation = 60)
plt.ylabel('수',fontsize=axis_label_size) 
plt.show()
```

## 지방소멸 위험은 사실일까?

* 지방소멸위험지수란 한 지역의 20∼39세 여성 인구를 65세 이상 인구로 나눈 값으로, 지수가 0.5 미만이면 인구소멸 위험 지역, 0.2 미만이면 인구소멸 고위험 지역으로 간주

### [그림5] 2020 시도별 지방소멸 위험지수

```python
city = pd.read_csv('C://Users/user/Desktop/통합데이터지도공모전/활용데이터/소멸위험지수.csv', encoding='ansi')
city
```

```python
# 위험정도 등급화
city['위험정도'] = 0
city.loc[city['지방소멸위험지수'] >= 0.5, '위험정도'] = '안전'
city.loc[(city['지방소멸위험지수'] <0.5) & (city['지방소멸위험지수']>=0.2), '위험정도'] = '위험'
city.loc[city['지방소멸위험지수'] < 0.2, '위험정도'] = '고위험'
```

```python
city
```
